```python
df = spark.table("csv.chat_gpt_data")
df.show()
```

```python
from pyspark.sql.functions import to_date, datediff

## Adding new column days
df = df.withColumn("days", datediff("End Date", "Start Date"))
df.show()
```

```python
from pyspark.sql.functions import regexp_replace, col, trim

df = df.withColumn(
    "Cost",
    trim(
        regexp_replace(
            regexp_replace("Cost", r"[\$,]", ""),  # Remove $ and commas
            r"\s+", ""                            # Remove all whitespace
        )
    )
)
df = df.withColumn(
    "Cost",
    col("Cost").cast("double")
)
display(df)
```
