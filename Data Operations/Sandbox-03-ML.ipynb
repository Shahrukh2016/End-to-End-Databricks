```python
df = spark.table('workspace.csv.bitcoin')

df.show()
```

```python
## Convert to pandas
pdf = df.toPandas()
pdf.head(2)
```

```python
pdf['snapped_at'] = pd.to_datetime(pdf['snapped_at'])
pdf.info()
```

```python
## Renaming column
pdf.rename(columns={'snapped_at': 'date'}, inplace=True)
pdf.head(2)
```

```python
## Converting market_cap into scientific value
pdf['market_cap'] = pdf['market_cap'].apply(
    lambda x: '{:,}'.format(float(x))
)
pdf.head(2)
```
